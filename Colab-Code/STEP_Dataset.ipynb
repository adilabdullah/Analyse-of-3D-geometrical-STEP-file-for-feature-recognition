```python
import csv
import pandas as pd
import xlsxwriter

f = open('famous.stp','r')
message=f.read();
data=message[message.find('DATA;')+5:]


stp=data[:data.find('ENDSEC;')]
stp=stp.replace(' ', '');
myList=stp.split(';');
###print(stp);
elemList=[];
refDicts={};
paramDicts={};
count=0;


refParamList=[];
refParamDict={};

for i in myList:

    refParamList=[];
    elemList.append(i[i.find('=')+2:i.find('(')]);
    refDicts[i[i.find('#')+1:i.find('=')]] = i[i.find('=')+1:i.find('(')];
    params=i[i.find('(')+1:i.find(';')].replace(")","").replace("(","");
    paramsList=params.split(',');
    for j in paramsList:
        if('#' in j):
            refParamList+=j.split('#');
    paramDicts[i[i.find('#')+1:i.find('=')]]=paramsList;
    refParamDict[i[i.find('#')+1:i.find('=')]]=list(filter(None, refParamList));

finalDict={};

filename = "Step_datasets_1.csv"
workbook = xlsxwriter.Workbook(filename)
worksheet = workbook.add_worksheet()
row=1;
#with open(filename, 'w') as file:
#    csv_writer = csv.writer(file,delimiter=',')
worksheet.write(0, 0,'KEY')
worksheet.write(0, 1,'ENTITY')
worksheet.write(0, 2,'D0')
worksheet.write(0, 3,'D1')
worksheet.write(0, 4,'D2')
worksheet.write(0, 5,'D3')
worksheet.write(0, 6,'D4')
worksheet.write(0, 7,'D5')
worksheet.write(0, 8,'D6')
worksheet.write(0, 9,'D7')
worksheet.write(0, 10,'REFERENCES')
for (key,value),(key1,value1),(key2,value2) in zip(paramDicts.items(),refDicts.items(),refParamDict.items()):
    print("\nKey::"+key+":Value::"+value1);
     #   csv_writer.writerow([key,value]);
 #   finalDict[key]={"Entity:":value1,"Elements":[]};
    col=2;
    worksheet.write(row, 0,key)
    worksheet.write(row, 1,value1)
    worksheet.write(row, 10,str(value2))
    for v in value:
        if(v[0:1]=='#'):
            print(refDicts.get(v[1:]),end=",");
            worksheet.write(row, col,refDicts.get(v[1:]))

         #   finalDict["Elements"]+=[refDicts.get(v[1:])];
        else:
            print(str(v),end=",");
            worksheet.write(row, col,str(v))
        col+=1;

          #  finalDict["Elements"]+=[str(v)];
    row+=1;

workbook.close()
```
